```python
import mysql.connector as sql

import admin_module as ad
import teacher_module as th
import student_module as stud

mydb=sql.connect(
    host="localhost",
    user="root",
    passwd="Root",
    database="college_mgmt"
    )

commit_handler=mydb.cursor(buffered=True)

def validate_user(privilege,user_name,password):
    success=False
    if privilege==1:
        commit_handler.execute("SELECT COUNT(*) CNT FROM USERS_TABLE WHERE NAME='%s' AND SET_PWD='%s' AND REGISTERED_AS='A' " %(user_name,password))
        rows=commit_handler.fetchall()
        if rows[0][0] >= 1:
            success=True
            print("Welcome {} !!!,Login successfully as Admin".format(user_name))
        else:
            print("Invalid username or password, user not found")
    elif privilege==2:
        commit_handler.execute("SELECT COUNT(*) CNT FROM USERS_TABLE WHERE NAME='%s' AND SET_PWD='%s' AND REGISTERED_AS='T' " %(user_name,password))
        rows=commit_handler.fetchall()
        if rows[0][0] >= 1:
            success=True
            print("Welcome {} !!!,Login successfully as Teacher".format(user_name))
        else:
            print("Invalid username or password, user not found")
    elif privilege==3:
        commit_handler.execute("SELECT COUNT(*) CNT FROM USERS_TABLE WHERE NAME='%s' AND SET_PWD='%s' AND REGISTERED_AS='S' " %(user_name,password))
        rows=commit_handler.fetchall()
        if rows[0][0] >= 1:
            success=True
            print("Welcome {} !!!,Login successfully as Student".format(user_name))
        else:
            print("Invalid username or password, user not found")
        
    return success


def get_user_details(option):
    login_option=option
    user_name=input("Enter a username: ") 
    password=input("Enter a password: ") 
    validation=validate_user(login_option,user_name,password)
    if validation==True:
        if login_option==1:
            obj1=ad.administration()
            obj1.admin()
        elif login_option==2:
            obj2=th.TeacherMode()
            obj2.teacher()
        elif login_option==3:
            objc3=stud.studentMode()
            objc3.student()
            
            
def main():
    print("Welcome to IIT")
    print("")
    print("Select the option from the below menu: \n1.Login as Admin \n2.Login as Teacher \n3.Login as Student")
    user_option=int(input())
    get_user_details(user_option)
    
    
main()   
```
